### 8) A partir du dataframe, créez le tableau d'aggrégation suivant : la moyenne des variables 'ENGINESIZE' et 'FUELCONSUMPTION_CITY' en fonction de la variable 'VEHICLECLASS'

### 9) Affichez la consommation moyenne par marque de voiture 

```python
conso_par_marque = df.groupby('MAKE')['FUELCONSUMPTION_COMB'].mean()
print(conso_par_marque.sort_values())
```
