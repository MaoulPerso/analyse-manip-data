# Solution de l'exercice : manipulation d'un fichier CSV avec pandas

```python
# import de la librairie pandas
import pandas as pd
```

## 1. Création d'un fichier CSV fictif (ventes.csv) à partir d'un DataFrame pandas

```python
data = {
    "Date": ["2025-01-01", "2025-01-01", "2025-01-02", "2025-01-02", "2025-01-03"],
    "Vendeur": ["Jean", "Marie", "Jean", "Marie", "Paul"],
    "Produit": ["Ordinateur", "Souris", "Clavier", "Ordinateur", "Écran"],
    "Quantité": [3, 10, 5, 2, 4],
    "Prix_unitaire": [500, 20, 30, 500, 200]
}

df_initial = pd.DataFrame(data)
df_initial.to_csv("ventes.csv", index=False)
```

## 2. Lecture du fichier CSV avec pandas

```python
def lire_csv(fichier):
    return pd.read_csv(fichier)

df = lire_csv("ventes.csv")
```

```python
df
```

## 3. Calcul du total des ventes

```python
def calcul_total_ventes(df):
    df['Total'] = df['Quantité'] * df['Prix_unitaire']
    return df['Total'].sum()

total_ventes = calcul_total_ventes(df)
print("Total des ventes :", total_ventes)
```

```python
df
```

## 4. Quantité totale vendue par chaque vendeur

```python
def quantite_par_vendeur(df):
    return df.groupby('Vendeur')['Quantité'].sum()

ventes_par_vendeur = quantite_par_vendeur(df)
print("Quantité totale vendue par vendeur :")
print(ventes_par_vendeur)
```

## 5. Produit le plus vendu

```python
def produit_plus_vendu(df):
    ventes_par_produit = df.groupby('Produit')['Quantité'].sum()
    print(ventes_par_produit)
    produit_max = ventes_par_produit.idxmax()
    quantite_max = ventes_par_produit.max()
    return produit_max, quantite_max

produit_max, quantite_max = produit_plus_vendu(df)
print(f"Produit le plus vendu : {produit_max} ({quantite_max} unités)")
```

## 6. Journée avec le plus de revenus

```python
def journee_plus_rentable(df):
    df['Total'] = df['Quantité'] * df['Prix_unitaire']
    revenus_par_jour = df.groupby('Date')['Total'].sum()
    print(revenus_par_jour)
    jour_max = revenus_par_jour.idxmax()
    revenus_max = revenus_par_jour.max()
    return jour_max, revenus_max

jour_max, revenus_max = journee_plus_rentable(df)
print(f"Journée avec le plus de revenus : {jour_max} (total : {revenus_max})")
```
