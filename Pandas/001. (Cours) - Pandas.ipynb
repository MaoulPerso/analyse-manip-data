## 6. Gérer les Données Manquantes

```python
# Vérifier les valeurs manquantes
df.isna().sum()
```

```python
# Remplacer les valeurs manquantes
df.fillna(0)
```

```python
# Supprimer les lignes avec des valeurs manquantes
df.dropna(inplace=True)
```

## 7. Groupement et Agrégation

```python
# Grouper par une colonne et agréger
result = df.groupby('name')['overall'].mean()
```

```python
result
```

```python
# Agrégations multiples
df.groupby('team').agg({'overall': ['mean', 'median']})
```

## 8. Fusion et Jointure

```python
data1 = pd.DataFrame({'Nom': ['Alice', 'Bob'], 'Ville': ['Paris', 'Lyon']})
data2 = pd.DataFrame({'Nom': ['Alice', 'Bob'], 'Score': [95, 80]})
print(data1)
print(data2)
```

```python
# Jointure interne
df_merge = pd.merge(data1, data2, on='Nom', how='inner')
```

```python
df_merge
```

### Autres formats de d'enregistrements

```python
df = pd.read_csv('../data/FIFA21Complete.csv', sep=";")
df.head()
```

```python
# -------------------------------------
# 2. Sauvegarde en Excel
# -------------------------------------
df.to_excel('data.xlsx', index=False, engine='openpyxl')  # Format .xlsx
```

```python
# -------------------------------------
# 3. Sauvegarde en JSON
# -------------------------------------
df.to_json('data.json', orient='records', lines=True)  # Lignes séparées
```

```python
# -------------------------------------
# 4. Sauvegarde en Pickle (sérialisation Python)
# -------------------------------------
df.to_pickle('data/data.pkl')
```

```python
# -------------------------------------
# 5. Sauvegarde en Parquet (optimisé pour gros volumes)
# -------------------------------------
df.to_parquet('data.parquet', engine='pyarrow', index=False)
```

```python
# -------------------------------------
# 6. Sauvegarde en Feather (rapide et léger)
# -------------------------------------
df.to_feather('data.feather')
```

```python
# -------------------------------------
# 7. Sauvegarde en HTML (utile pour visualisation web)
# -------------------------------------
df.to_html('data.html', index=False)
```
