```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
sns.set_style('whitegrid')
sns.displot(df['fare'], kde=False, color='red', bins=30)
```

```python
sns.boxplot(x='class', y='age', data=df, palette='Blues_d')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
sns.set_style('whitegrid')
sns.swarmplot(x='class', y='age', data=df, palette='Set1')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
sns.set_style('whitegrid')
sns.jointplot(x='fare', y='age', data=df)
```
