## Exercice 1 : Tracer un graphique simple

Dans cet exercice, nous allons tracer un graphique simple de type `ligne` en utilisant `matplotlib`.

### Instructions :
1. Importez la bibliothèque `matplotlib.pyplot`.
2. Créez un graphique en utilisant les données suivantes :
   - x = [1, 2, 3, 4, 5]
   - y = [2, 3, 5, 7, 11]
3. Affichez le graphique avec `plt.show()`.

```python
import matplotlib.pyplot as plt

# Données
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Tracer le graphique
plt.plot(x, y)
plt.show()
```

## Exercice 2 : Personnalisation du graphique

Dans cet exercice, vous allez personnaliser un graphique.

### Instructions :
1. Tracez un graphique en utilisant les mêmes données que l'exercice précédent.
2. Ajoutez un titre à votre graphique, une légende, et des labels pour les axes x et y.
3. Changez la couleur de la ligne en rouge et ajoutez des points marqués sur la courbe.

Utilisez les fonctions suivantes :
- `plt.title()` pour ajouter un titre.
- `plt.xlabel()` et `plt.ylabel()` pour les labels des axes.
- `plt.plot()` avec des arguments pour la couleur et les marqueurs.

```python
# Tracer un graphique personnalisé
plt.plot(x, y, color='red', marker='o')

# Personnalisation
plt.title('Graphique des nombres premiers')
plt.xlabel('x')
plt.ylabel('y')

# Afficher la légende
plt.legend(['Prime Numbers'])

# Affichage
plt.show()
```

## Exercice 3 : Sous-graphiques (Subplots)

Dans cet exercice, vous allez travailler avec des sous-graphiques.

### Instructions :
1. Créez deux graphiques côte à côte en utilisant la fonction `plt.subplot()`.
2. Le premier graphique sera une courbe de type `line` avec les données de l'exercice 1.
3. Le second graphique sera un graphique à barres avec les mêmes données `x` et `y`.
4. Ajoutez des titres à chaque sous-graphe.

```python
# Créer deux sous-graphiques
plt.subplot(1, 2, 1)  # 1 ligne, 2 colonnes, premier graphique
plt.plot(x, y, color='blue')
plt.title('Graphique linéaire')

plt.subplot(1, 2, 2)  # 1 ligne, 2 colonnes, deuxième graphique
plt.bar(x, y, color='green')
plt.title('Graphique à barres')

# Affichage
plt.tight_layout()
plt.show()
```

## Exercice 4 : Graphique avec plusieurs axes

Dans cet exercice, vous allez apprendre à utiliser plusieurs axes pour représenter différentes données.

### Instructions :
1. Créez un graphique avec deux axes verticaux. 
2. Le premier axe représentera les données `y` de l'exercice 1.
3. Le deuxième axe représentera une autre fonction comme `y2 = [3, 6, 9, 12, 15]`.
4. Les deux axes doivent avoir des étiquettes distinctes.

```python
# Créer un graphique avec deux axes
fig, ax1 = plt.subplots()

# Tracer les données sur le premier axe
ax1.plot(x, y, 'r-', label='Prime Numbers')
ax1.set_xlabel('x')
ax1.set_ylabel('y (Prime Numbers)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Créer un deuxième axe partageant le même x
ax2 = ax1.twinx()
y2 = [3, 6, 9, 12, 15]
ax2.plot(x, y2, 'b-', label='Second Function')
ax2.set_ylabel('y2 (Other Function)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Affichage
plt.title('Graphique avec axes multiples')
plt.show()
```
