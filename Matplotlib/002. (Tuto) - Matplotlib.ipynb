```python
# Importation de l'interface pyplot de Matplotlib

import matplotlib.pyplot as plt
```

# Tracés linéaires

```python
# Tracé d'une courbe (line plot)
annee = [1950, 1970, 1990, 2010]
population = [2600000, 3700000, 5300000, 7000000]

plt.plot(annee, population)
plt.xlabel("Année")
plt.ylabel("Population (millions)")
plt.title("Evolution de la population")
plt.show()
```

```python
# On peut personnaliser le style de ligne, le marqueur et la couleur
plt.plot(annee, population, linestyle='--', marker='v', color='r')
plt.xlabel("Année")
plt.ylabel("Population (millions)")
plt.title("Evolution de la population")
plt.show()
```

```python
# Importez le fichier
gap = pd.read_csv("gapminderDataFiveYear.txt", sep='\t')

# Affichez les cinq premières lignes de gap

gap.head(15)
```

```python
gap.country.unique()
```

```python
gap[gap.country == 'France']
```

```python
df_1992 = gap[gap['year'] == 1992]
df_1992
```

```python
# Tracez un nuage de points montrant l'espérance de vie en fonction du PIB
# On prend que l'année 1992
plt.figure(figsize = (12, 7))

plt.scatter(df_1992['gdpPercap'], df_1992['lifeExp'])

plt.xlabel("PIB")

plt.ylabel("Espérance de vie)")

plt.title("Espérance de vie VS PIB")

plt.show()
```

```python
# Reprenez le code du graphique précédent en ajoutant plt.xscale('log') et plt.grid(True)

plt.figure(figsize = (12, 7))

plt.scatter(df_1992['gdpPercap'], df_1992['lifeExp'])

plt.xscale('log')

plt.xlabel("Log du PIB")

plt.ylabel("Espérance de vie)")

plt.title("Espérance de vie VS PIB")

plt.grid(True)

plt.show()
```

```python
gap.continent.unique()
```

```python
# Tracez un nuage de points montrant l'espérance de vie en fonction du PIB dans le continent africain

afr = gap[gap.continent == "Africa"]

plt.figure(figsize = (12, 7))

plt.scatter(afr['gdpPercap'], afr['lifeExp'])

plt.xlabel("PIB")

plt.ylabel("Espérance de vie)")

plt.title("Espérance de vie VS PIB")

plt.show()
```

# Diagramme à barres

```python
# Exemple de diagramme à barre

x1 = [1, 3, 4, 5, 6, 7, 9]

y1 = [4, 7, 2, 4, 7, 8, 3]

plt.bar(x1, y1, label="Blue Bar", color='b')

plt.xlabel("nombre")

plt.ylabel("Taille")

plt.title("Exemple de diagramme à barre")

plt.legend()

plt.show()
```

```python
# Créez une dataframe du PIB moyen par continent en 1992

pib_moy = df_1992.groupby("continent")['gdpPercap'].mean()
pib_moy = pd.DataFrame(pib_moy)
pib_moy
```

```python
# Construire un diagramme à barres montrant le PIB moyen par continent en 1992

plt.bar(pib_moy.index, pib_moy['gdpPercap'], label = "PIB par continent", color = 'g')

plt.xlabel("Continent")

plt.ylabel("PIB moyen")

plt.title("PIB moyen par continent")

plt.legend()

plt.show()
```

# Diagramme circulaire

```python
# Exemple de diagramme circulaire
labels = "S1", "S2", "S3"
sections = [56, 66, 24]
plt.pie(sections, labels = labels, colors = ['c', 'g', 'y'])
plt.title("Exemple de graphique circulaire")
plt.show()
```

```python
# On peut customiser le graphique précédent
# Créer un graphique avec fond noir
plt.figure(figsize=(7, 7), facecolor='black')
plt.pie(sections, labels = labels, colors = ['c', 'g', 'y'],
        explode = (0.1, 0.1, 0.1), # Essayez aussi explode = (0, 0.1, 0)
        autopct = '%1.2f%%')

plt.title('Exemple de diagramme circulaire', color='white') 

plt.show()
```

```python
# Construisez un diagramme circulaire montrant le PIB moyen par continent

plt.pie(pib_moy.gdpPercap, labels = pib_moy.index)
plt.title("PIB moyen par continent en 1992")
plt.show()
```

# Histogramme

```python
# Construction d'un histogramme

plt.hist(apple.Close)
plt.title("Histogramme du prix de cloture de APPLE")
plt.show()
```

```python
# Changement du nombre de bacs (bins)

plt.hist(apple.Close, bins = 50)
plt.title("Histogramme du prix de cloture de APPLE")
plt.show()
```

```python
# Avec une boucle for, construire les histogrammes des variables population, life_exp et gdp_cap

for i in ["pop", "lifeExp", "gdpPercap"]:
  plt.hist(df_1992[i])
  plt.title(i)
  plt.show()
```

# Subplots et Sauvegarde d'un graphique

```python
plt.subplot(2, 1, 1)
plt.hist(apple.Close)
plt.ylabel("Prix de cloture")
plt.title("Histogramme")

plt.subplot(2, 1, 2)
plt.boxplot(apple.Close, vert=False)
plt.yticks([1], ["Prix de cloture"])
plt.title("Boite à moustache")

plt.tight_layout()
plt.savefig("AppleClosedPrice.png", dpi=900)
```

```python
noms_variables = [
    "age", "workclass", "fnlwgt", "education", "education-num",
    "marital-status", "occupation", "relationship", "race",
    "sex", "capital-gain", "capital-loss", "hours-per-week",
    "native-country", "class"
]

df = pd.read_csv(
    "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",
    header=None,
    names = noms_variables
)

df.head()
```

```python
df.dtypes
```

```python
# Colonne numériques
cols_num = df.select_dtypes(exclude="object").columns.to_list()
cols_num
```

```python
plt.figure(figsize=(18,7)) 
for i,col in enumerate(cols_num):
    plt.subplot(2, len(cols_num), i+1)
    plt.hist(df[col])
    plt.title(col)
for i,col in enumerate(cols_num):
    plt.subplot(2, len(cols_num), i+len(cols_num)+1)
    plt.boxplot(df[col], vert=False)
    plt.yticks([])
plt.tight_layout()
plt.savefig('Visualisation.png', dpi=900)
```
