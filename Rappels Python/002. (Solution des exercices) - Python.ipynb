## Exercice 1 : Manipulation de listes

Écrivez une fonction `carres_pairs(liste)` qui prend en entrée une liste de nombres et retourne une nouvelle liste contenant les carrés des nombres pairs uniquement.

Exemple :
```python
carres_pairs([1, 2, 3, 4, 5, 6])
```
Résultat attendu : `[4, 16, 36]`


### Correction

````python
def carres_pairs(liste):
    return [x**2 for x in liste if x % 2 == 0]

# Test de la fonction
print(carres_pairs([1, 2, 3, 4, 5, 6]))
````

## Exercice 2 : Programmation orientée objet

Implémentez une classe `CompteBancaire` représentant un compte bancaire avec :
- Un attribut `solde` (initialisé à 0 par défaut).
- Une méthode `depot(montant)` pour ajouter de l'argent.
- Une méthode `retrait(montant)` pour retirer de l'argent (si le solde est suffisant).
- Une méthode `afficher_solde()` qui affiche le solde actuel.

Exemple d'utilisation :
```python
compte = CompteBancaire()
compte.depot(100)
compte.retrait(50)
compte.afficher_solde()
```
Résultat attendu : `Solde actuel : 50`

### Correction

````python
class CompteBancaire:
    def __init__(self):
        self.solde = 0

    def depot(self, montant):
        self.solde += montant

    def retrait(self, montant):
        if montant <= self.solde:
            self.solde -= montant
        else:
            print("Fonds insuffisants.")

    def afficher_solde(self):
        print(f"Solde actuel : {self.solde}")

# Test de la classe
compte = CompteBancaire()
compte.depot(100)
compte.retrait(50)
compte.afficher_solde()
````
