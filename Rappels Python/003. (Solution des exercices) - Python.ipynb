Nous allons créer un fichier donnees.csv pour l'exercice en utilisant cette fois la librairie csv.
Vous pouvez exécuter la cellule ci-dessous pour créer le fichier que nous allons utiliser pour l'exercice.

```python
import csv

# Données à écrire dans le fichier CSV
donnees = [
    ["Nom", "Âge", "Ville"],
    ["Alice", 25, "Paris"],
    ["Bob", 30, "Lyon"],
    ["Charlie", 22, "Marseille"],
    ["David", 35, "Bordeaux"],
    ["Emma", 28, "Nice"]
]

# Création du fichier CSV
with open("donnees.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(donnees)

print("Fichier donnees.csv créé avec succès !")
```

## Exercice 2 : Extraction d'une colonne spécifique

Écrivez un script qui extrait uniquement les noms des personnes contenues dans `donnees.csv`.

### Correction

```python
with open('donnees.csv', 'r', encoding='utf-8') as f:
    lignes = f.readlines()
    for ligne in lignes[1:]:  # On saute l'en-tête
        colonnes = ligne.strip().split(',')
        print(colonnes[0])  # La première colonne est le nom
```
