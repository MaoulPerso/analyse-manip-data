*      Simplicité du langage

## **3.1 Les variables**

En Python, la déclaration d'une variable et son initialisation se font en même temps :

```python
x = 1
```

On affecte à la variable x la valeur 1. Une variable est caractérisée par :
* **un identificateur** : il peut contenir des lettres, des chiffres, des blancs soulignés mais il ne peut commencer par un chiffre. Python est sensible à la casse.

### **3.2.10 Les compréhension de listes, de dictionnaires et de sets**


```python
l = [1, 2, 2, 4, 5, 14, 21, 1]

b = []
for x in l:
    if x > 5:
        b.append(x)
print(b)
```

```python
l = [1, 2, 2, 4, 5, 14, 21, 1]
b = [x for x in l if x > 5] # Avec les compréhensions de list
print(b)
```

# **4. Boucles et conditions**

## **4.1 Boucle for**

```python
for x in range(10):
    print(x**2)
```

## **4.2 Boucle while**

```python
i = 0
impairs = []
while i < 20:
    if i%2 == 1:
        impairs.append(i)
    i = i + 1
print(impairs)
```

## **4.3 Condition if/else**

```python
age = 14

if age > 16:
    print("Vous avez plus de 16 ans.")
elif age < 0:
    print("Tu te moque de moi?")
else:
    print("Tu es encore jeune.")
```

```python
def est_premier(n):
    for d in range(2, n//2+1):
        if n%d==0:
            return False
    return True
```

```python
x = 25
y = 23
print(x, "est un nombre premier? ", est_premier(x))
print(y, "est un nombre premier? ", est_premier(y))
```

# **6. Manipulation de fichiers avec Python**

Les fichiers servent à stocker des données de manière durable.


```python
noms_joueurs = ["Antoine Dupont", "Grégory Alldritt", "Romain Ntamack"]

nom_fichier = "joueurs_france_rugby.txt"

with open(nom_fichier, "w", encoding="utf-8") as fichier:
    for joueur in noms_joueurs:
        fichier.write(joueur + "\n")

print(f"Le fichier '{nom_fichier}' a été créé avec succès !")
```

```python
nom_fichier = "joueurs_france_rugby.txt"

with open(nom_fichier, "r") as fichier:
    contenu = fichier.readlines()

contenu = [ligne.strip() for ligne in contenu]
print("Contenu du fichier :", contenu)
```

```python
with open(nom_fichier, "r", encoding="utf-8") as fichier:
    joueurs = fichier.readlines()

joueurs = [ligne.strip() for ligne in joueurs]
print("Contenu du fichier :", joueurs)
```

# **7. Exceptions en Python**

Une exception est une erreur qui se produit lors de l'exécution du programme.

```python
def div(a, b):
    return a/b

print(div(5,2))
#print(div(5,0))
```

```python
def div(a, b):
    try:
        return a/b
    except ZeroDivisionError:
        print("Attention, division par 0")

print(div(5,2))
print(div(5,0))
#print(div(5,'0'))
```

```python
def div(a, b):
    try:
        return a/b
    except ZeroDivisionError:
        print("Attention, division par 0")
    except TypeError:
        print("Il faut des int!")

print(div(5,2))
print(div(5,0))
print(div(5,'0'))
```

# **8. Classes et objets**

Une classe est un ensemble de variables et de fonctions qui définissent un objet.

```python
class MaClass:
    pass
```

```python
c = MaClass()
print(type(c))
```

La méthode *isinstance* permet de savoir si un objet est une instance d'une classe donnée.

```python
c = MaClass()
print(type(c))
print(isinstance(c, MaClass))
```

```python
class Joueurs:
    
    def __init__(self, prenom, nom):
        self.prenom = prenom
        self.nom = nom
        
    def setAge(self, age):
        self.age = age
    
    def getAge(self):
        return self.age
```

```python
e1 = Joueurs("Antoine", "Grégory")
print(e1.prenom, ' ', e1.nom)
e1.setAge(29)
print(e1.age)
print(e1.getAge())
```
