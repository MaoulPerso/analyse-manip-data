## Exercice 1 : Manipulation des chaînes de caractères
**Enoncé :**
Écrivez un programme qui demande une phrase à l'utilisateur et affiche cette phrase en inversant l'ordre des mots.
Pour rappel la fonction pour demander à l'utilisateur de renseigner une valeur est "input()"

```python
phrase = input("Entrez une phrase : ")
mots = phrase.split()
phrase_inverse = " ".join(mots[::-1])
print(phrase_inverse)
```

## Exercice 2 : Opérations sur les listes
**Enoncé :**
Créez une liste de nombres et trouvez la somme des éléments pairs.

```python
nombres = [12, 7, 5, 14, 9, 10, 22, 3]
somme_pairs = sum(n for n in nombres if n % 2 == 0)
print("Liste :", nombres)
print("Somme des nombres pairs :", somme_pairs)
```

## Exercice 3 : Génération de nombres premiers
**Enoncé :**
Écrivez une fonction qui génère tous les nombres premiers jusqu'à `n`.

```python
def est_premier(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def nombres_premiers(limit):
    return [n for n in range(2, limit+1) if est_premier(n)]

print(nombres_premiers(50))
```
